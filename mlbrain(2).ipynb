```python
df=pd.read_csv("/kaggle/input/brain-tumor/Brain Tumor.csv")
df.head(8)
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.replace([np.inf, -np.inf], np.nan)
df.dropna()
print (len(df))
#we dont need Image and Class cols
aux=df.iloc[:,2:]
scaler=MinMaxScaler()
datanp=scaler.fit_transform(aux)
print (datanp[1])
print (type(datanp))
#we want a df, not a nparray
data=pd.DataFrame(datanp, columns=aux.columns)
data.head()
tumor=df.Class
tumor.head()
print (type(tumor))
```

```python
tumor_classes=['0','1']
plt.bar(tumor_classes,tumor.value_counts())
```

```python
#Graphs meant to represent some of the correlations
pos=df['Class']== 1
neg=df['Class']== 0
#df_without_image=data.copy()
df_without_image=df.drop('Image',axis=1,inplace=False)

fig, axes = plt.subplots(5,1,figsize=(20,20))

sns.distplot(ax=axes[0],a = df_without_image[neg]['Energy'], label="No Tumor" )
sns.distplot(ax=axes[0],a= df_without_image[pos]['Energy'], label="Tumor")
sns.distplot(ax=axes[1],a = df_without_image[neg]['Mean'], label="No Tumor" )
sns.distplot(ax=axes[1],a= df_without_image[pos]['Mean'], label="Tumor")
sns.distplot(ax=axes[2],a = df_without_image[neg]['Variance'], label="No Tumor" )
sns.distplot(ax=axes[2],a= df_without_image[pos]['Variance'], label="Tumor")
sns.distplot(ax=axes[3],a = df_without_image[neg]['Homogeneity'], label="No Tumor" )
sns.distplot(ax=axes[3],a= df_without_image[pos]['Homogeneity'], label="Tumor")
sns.distplot(ax=axes[4],a = df_without_image[neg]['Skewness'], label="No Tumor" )
sns.distplot(ax=axes[4],a= df_without_image[pos]['Skewness'], label="Tumor")
plt.figure()
sns.catplot(data=df_without_image, kind="violin",x="Class",  y="Energy" )
plt.title("Energy vs Tumor")
plt.legend()
sns.catplot(data=df_without_image, kind="violin",x="Class",  y="Homogeneity" )
plt.title("Homogeneity vs Tumor")
plt.legend()

colors = {0:"blue", 1:"orange"}
fig, axes = plt.subplots(3,3)
grouped = pd.concat([tumor,data],axis=1).groupby("Class")
cols = ["Entropy", "Energy", "Homogeneity"]

for i in range(3):
    for j in range(3):
        for key, group in grouped:
            group.plot(ax=axes[i,j], kind="scatter", x=cols[i], y=cols[j], label=key, c=colors[key],figsize=(20,20))

plt.subplots(figsize=(15,15))
plt.title("Colourmap reprsenting correlations between given variables",fontsize=15)
sns.heatmap(data.corr(),linewidths=.05,cmap="bone")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data,tumor, train_size=0.75)
```

```python
print (X_test.head(10))
print(y_test.head(10))
```

```python
#LinReg
classifier= LinearRegression()
classifier.fit(X_train , y_train )
score=classifier.score(X_test,y_test)
print("The score for LingReg is: ", score," or ", score*100,"%")
```

```python
#LogReg
classifier= LogisticRegression(random_state=10)
classifier.fit(X_train , y_train )
score=classifier.score(X_test,y_test)
print("The score for LogReg is: ", score," or ", score*100,"%")
```

```python
#KNN

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
score = knn.score(X_test, y_test)
print("The score for KNN is: ", score," or ", score*100,"%")

```

```python
#DecisionTree
classifier=DecisionTreeClassifier()
classifier.fit(X_train,y_train)
score=classifier.score(X_test,y_test)
print("The score for DecisionTree is: ", score," or ", score*100,"%")
```

```python
#RandomForest
classifier=RandomForestClassifier(n_estimators=2000,random_state=10)
classifier.fit(X_train, y_train)
score = classifier.score(X_test, y_test)
print("The score for RFC is: ", score," or ", score*100,"%")
```
